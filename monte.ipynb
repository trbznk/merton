```python
import ctypes
```

```python
class Simulation(ctypes.Structure):
    _fields_ = [
        ("asset_corr", ctypes.c_float),
        ("cut_off", ctypes.c_float),
        ("EAD", ctypes.c_float),
        ("LGD", ctypes.c_float),
        ("pf_size", ctypes.c_int),
        ("n_sims", ctypes.c_int),
        ("L_PF", ctypes.POINTER(ctypes.c_float))
    ]

libmonte = ctypes.CDLL("./monte.so")
libmonte.simulate.argtypes = [ctypes.POINTER(Simulation)]
```

```python
sim = Simulation(0.05, -2.0, 1.0, 0.6, 100, 1000, (ctypes.c_float*1000)())
libmonte.simulate(ctypes.byref(sim))
for i in range(sim.n_sims):
    print(f"{sim.L_PF[i]=}")
```
